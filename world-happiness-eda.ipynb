```python
gdp_mean=df['gdp_per_capita'].mean()
score_mean=df['score'].mean()

high_gdp_low_happy = df[
    (df['gdp_per_capita'] > gdp_mean) &
    (df['score'] < score_mean)
][['country','region','gdp_per_capita','score']].sort_values(by='gdp_per_capita', ascending=False)

print(high_gdp_low_happy)
```

```python
# Bonus Visualization - Highlight Exceptions

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='gdp_per_capita', y='score', hue='region')
plt.axvline(gdp_mean, linestyle='--', color='red')
plt.axhline(score_mean, linestyle='--', color='blue')

plt.title("GDP vs Happiness (Quadrant Analysis)")
plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.show()
```
